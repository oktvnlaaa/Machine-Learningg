#Library

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import accuracy_score
```

# Dataset

```python
df2 = pd.read_csv ("https://raw.githubusercontent.com/rahmantaufik27/dataset/main/study/student_score_pass.csv")
df2
```

# Splitting Data/Prepocessing

```python
# pemilihian feature
df_training = df2.drop(['npm', 'pass_test'], axis=1)

# hapus NaN values
df_training = df_training.dropna()
df_training
```

# Modelling

```python
model = AgglomerativeClustering(n_clusters=4).fit(df_training)
# hasil clustering
df_training["kluster"] = model.labels_
df_training
```

# Accuracy Score

```python
# Initializing the arrays (multiclass case)

y_pred = [99,	71.0,	93.0,	52.0,	44.0,	24]

y_true = [86,	71.0,	84.0,	17.0,	21.0,	41]

# Calculating and printing the result
accuracy_score(y_true, y_pred, normalize=False)
```
